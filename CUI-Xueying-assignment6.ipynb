```python
import pandas as pd
import bqplot
import numpy as np
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.colors as mpl_colors
```

```python

#the left component:heat map
buildings = pd.read_csv('/Users/cuixueying/Desktop/UIUC courses/445 data visualization/week2/building_inventory.csv',
                       na_values = {'Square Footage': 0,
                       'Year Acquired': 0,
                       'Year Constructed': 0}
                       )
```

```python
#because 'Agency Name' is not a numerical variable and be difficult to used in building some plots
#so add a new numerical column to encoding every value in 'Agency Name' 
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
enc.fit(buildings['Agency Name'])
agency_index = enc.transform(buildings['Agency Name'])
buildings_agency_index = pd.concat([buildings, pd.DataFrame(agency_index,columns=['Agency Index'])],axis=1)
buildings_agency_index
```

```python
#delete rows with NaNs
buildings=buildings.dropna()
buildings
#to check the data
buildings_agency_index=buildings_agency_index.dropna()
buildings_agency_index#to check the data
```

```python
#build a heat map to check
hist2d, x_edges, y_edges = np.histogram2d(buildings_agency_index['Congress Dist'],
                                              buildings_agency_index['Agency Index'], 
                                              weights=buildings_agency_index['Square Footage'], 
                                              bins = [20,35])
```

```python
def generate_histogram_from_y_x(buildings_agency_index, nx=20, ny=35, xmin=0, xmax=20,
                                     ymin=0, ymax=35,
                                     takeLog=True):
    x_bins = np.linspace(xmin, xmax, nx+1)
    y_bins = np.linspace(ymin, ymax, ny+1)
    hist2d, x_edges, y_edges = np.histogram2d(buildings_agency_index['Congress Dist'],
                                              buildings_agency_index['Agency Index'], 
                                              weights=buildings_agency_index['Square Footage'], 
                                              bins = [x_bins,y_bins])
    hist2d = hist2d.T
    if takeLog:
        hist2d[hist2d <= 0] = np.nan 
        hist2d = np.log10(hist2d)
    x_centers = (x_edges[:-1] + x_edges[1:]) / 2
    y_centers = (y_edges[:-1] + y_edges[1:]) / 2
    return hist2d, x_centers, y_centers

hist2d, x_centers, y_centers = generate_histogram_from_y_x(buildings_agency_index)

# (1) add scales - colors, x & y
agent_name=list(enc.classes_)   

col_sc = bqplot.ColorScale(scheme="BrBG", 
                           min=np.nanmin(hist2d), 
                           max=np.nanmax(hist2d))
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

# (2) create axis - for colors, x & y
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'left')

x_ax = bqplot.Axis(scale = x_sc, label='Congress Dist')
y_ax = bqplot.Axis(scale = y_sc,
                   side = 'right',
                   label = 'Agency Index',
                   orientation = 'vertical',
                   )

heat_map = bqplot.GridHeatMap(color = hist2d,
                              row = y_centers, 
                              column = x_centers,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'},
                              anchor_style = {'fill':'blue'}, 
                              selected_style = {'opacity': 1.0},
                              unselected_style = {'opacity': 1.0})

fig = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig.layout.width = '600px'
fig.layout.height = '800px'
agency = pd.DataFrame(agent_name,index=list(range(1,36)))
print(agency)
fig
```

```python


# (I) CREATE LABEL
mySelectedLabel = ipywidgets.Label()

# (II) HEAT MAP

# (1) add scales - colors, x & y
col_sc = bqplot.ColorScale(scheme='Accent', 
                           min=np.nanmin(hist2d), 
                           max=np.nanmax(hist2d))
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

# (2) create axis - for colors, x & y
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')

x_ax = bqplot.Axis(scale = x_sc, label='Congress Dist')
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical', 
                   label = 'Agency Index')

# (3) Marks
heat_map = bqplot.GridHeatMap(color = hist2d,
                              row = y_centers, 
                              column = x_centers,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'},
                              anchor_style = {'fill':'blue'}, 
                              selected_style = {'opacity': 1.0},
                              unselected_style = {'opacity': 1.0})
```

```python
#line plots
# (1-2) scales & ax in usual way
x_scl = bqplot.LinearScale() # note we are back to linears
y_scl = bqplot.LinearScale()
ax_xcl = bqplot.Axis(label='Year Acquired', scale=x_scl)
ax_ycl = bqplot.Axis(label='Total Square Footage', scale=y_scl, 
                    orientation='vertical', side='left')

# create the data mask for each binned region like we did before:
i,j = 19,0
xside = [x_edges[j], x_edges[j+1]]
yside = [y_edges[i],y_edges[i+1]]
mask = ( (buildings_agency_index['Agency Index'] >= yside[0]) & (buildings_agency_index['Agency Index'] <=yside[1]) &\
                (buildings_agency_index['Congress Dist'] >= xside[0]) & (buildings_agency_index['Congress Dist']<=xside[1]) )
```

```python
# line plot, weight by duration, 10 bins in years:
tsf, tsf_edges = np.histogram(buildings_agency_index['Year Acquired'][mask],
                              weights=buildings_agency_index['Square Footage'][mask],
                              bins=10)
tsf_centers = (tsf_edges[:-1] + tsf_edges[1:]) / 2
tsf_line = bqplot.Lines(x=tsf_centers, y=tsf, 
                          scales={'x':x_scl, 'y':y_scl})
```

```python
def get_data_value(change):
    if len(change['owner'].selected) == 1: #only 1 selected
        i,j = change['owner'].selected[0]
        v = hist2d[i,j] # grab data value
        mySelectedLabel.value = 'Total Square Footage = ' + str(v) # set our label
        # Histogram:
        xside = [x_edges[j], x_edges[j+1]]
        yside = [y_edges[i],y_edges[i+1]]
        mask = ( (buildings_agency_index['Agency Index'] >= yside[0]) & (buildings_agency_index['Agency Index'] <=yside[1]) &\
                 (buildings_agency_index['Congress Dist'] >= xside[0]) & (buildings_agency_index['Congress Dist']<=xside[1]) )
    if len(buildings_agency_index['Year Acquired'][mask]) > 0: # make sure point exist so no histogram errors!
            tsf, tsf_edges = np.histogram(buildings_agency_index['Year Acquired'][mask],
                                          weights=buildings_agency_index['Square Footage'][mask],
                                          bins=10)
            tsf_centers = (tsf_edges[:-1] + tsf_edges[1:]) / 2
            tsf_line.x = tsf_centers
            tsf_line.y = tsf

# make sure we connect to heatmap     
heat_map.observe(get_data_value, 'selected')
```

```python
fig_heatmap = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig_tsf = bqplot.Figure(marks = [tsf_line], axes = [ax_xcl, ax_ycl])
fig_heatmap.layout.min_height= '800px'
fig_tsf.layout.min_height = '800px'
fig_heatmap.layout.min_width = '500px'
fig_tsf.layout.min_width = '500px'

plots = ipywidgets.HBox([fig_heatmap,fig_tsf])
myDashboard = ipywidgets.VBox([mySelectedLabel, plots])
print(agency)
myDashboard
```
